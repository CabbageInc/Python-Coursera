```python
for i in range(1, 11):
    for j in range(1, 11):
        print(i, "*", j, "=", i * j, end="\t") # Print the equation
    print() # Move to the next line after each row
```

```python
numbers = [3, 9, 1, 10, 5, 2, 8]

for number in numbers:
    if number%2==0:
        print(number, "is even")
    else:
        print(number, "is odd")
```
